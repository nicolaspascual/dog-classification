```python
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt; plt.rcParams["figure.figsize"] = 20,20
import seaborn as sns
import pandas as pd
```

```python
heights_train = []
widths_train = []
for im in glob('../data/train/**/*.jpg'):
    width, height = Image.open(im).size
    heights_train.append(height)
    widths_train.append(width)
```

```python
heights_val = []
widths_val = []
for im in glob('../data/validation/**/*.jpg'):
    width, height = Image.open(im).size
    heights_val.append(height)
    widths_val.append(width)
```

```python
heights_test = []
widths_test = []
for im in glob('../data/test/**/*.jpg'):
    width, height = Image.open(im).size
    heights_test.append(height)
    widths_test.append(width)
```

```python
ax = plt.subplot(3,1,1)
sns.distplot(heights_train, label='Height Train')
sns.distplot(widths_train, label='Width Train')
ax.set_xlim((0, 1000))
plt.legend()

ax = plt.subplot(3,1,2)
sns.distplot(heights_val, label='Height Validation')
sns.distplot(widths_val, label='Width Validation')
ax.set_xlim((0, 1000))
plt.legend()

ax = plt.subplot(3,1,3)
sns.distplot(heights_test, label='Height Test')
sns.distplot(widths_test, label='Width Test')
ax.set_xlim((0, 1000))
plt.legend()
```

```python
ax = plt.subplot(2,1,1)
sns.distplot(heights_train, label='Height Train')
sns.distplot(heights_val, label='Height Validation')
sns.distplot(heights_test, label='Height Test')
ax.set_xlim((0, 1000))
plt.legend()

ax = plt.subplot(2,1,2)
sns.distplot(widths_train, label='Width Train')
sns.distplot(widths_val, label='Width Validation')
sns.distplot(widths_test, label='Width Test')
ax.set_xlim((0, 1000))
plt.legend()
```

```python
pd.DataFrame(
    [heights_train, heights_val, heights_test, widths_train, widths_val, widths_test],
    index=['heights_train', 'heights_val', 'heights_test', 'widths_train', 'widths_val', 'widths_test']
).T.describe()
```
